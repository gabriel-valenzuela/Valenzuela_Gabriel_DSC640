# Python Plots

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import squarify
import numpy as np
from scipy.stats import kde
import squarify
import seaborn as sns
```

```python
# Import Data
costco_df = pd.read_csv('costcos-geocoded.csv')
costco_df.head()
```

```python
# Import Data
nba_df = pd.read_csv('ppg2008.csv')
nba_df.head()
```

## Heat Map

```python
squarify.plot(sizes=nba_df['PTS'], label=nba_df['MIN'], alpha=0.8)
plt.title('Points by Minutes of NBA Players in 2008')
plt.show()
```

## Spatial Chart

```python
BBox = ((costco_df['Longitude'].min(), costco_df['Longitude'].max(), 
        costco_df['Latitude'].min(), costco_df['Latitude'].max()))
print(BBox)
```

```python
ruh_m = plt.imread('map.png')
```

```python
fig, ax = plt.subplots(figsize = (8,7))
ax.scatter(costco_df['Longitude'], costco_df['Latitude'], zorder=1, alpha=0.2, c='b', s=10)

ax.set_title('Plotting Costco Spatial Data of U.S. Map')
ax.set_xlim(BBox[0], BBox[1])
ax.set_ylim(BBox[2], BBox[3])

#ax.imshow(ruh_m, zorder = 0, extent = BBox, aspect = 'equal')
```

## Contour Plot

```python
[X,Y] = np.meshgrid(nba_df['PTS'], nba_df['MIN'])
fig, ax = plt.subplots(1,1)
Z = np.cos(X/2) + np.sin(Y/4)
ax.contour(X,Y,Z)
ax.set_title('Contour Plot')
ax.set_xlabel('NBA Pts')
ax.set_ylabel('NBA Mins')
plt.show()
```
